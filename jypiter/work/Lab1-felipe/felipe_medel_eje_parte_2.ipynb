```python
# 1.- Leer el archivo Countries.csv en un dataframe
import pandas as pd
f = open('Countries.csv','r')
countries = pd.read_csv(f)
countries
```

```python
#Imprimir una lista de los países que no tienen linea costerá
# y no estan en la EU, junto con sus puntos más altos, 
# ordene de menor a mayor sus puntos más altos.
print(countries[(countries.coastline == 'no') & (countries.EU == 'no')].sort_values(by='highest') ) #python 3
#print countries[(countries.coastline == 'no') & (countries.EU == 'no')].sort_values(by='highest')  #python 2
```

```python
# Encontrar el punto medio más alto de los países en la UE y no en la UE,
# Encontrar el punto medio más alto de los países con y sin línea costera
# Nota: Cuando hay más de una "consulta" sólo se ve el último resultado.
# Intente usar print
# Consejo: ¡Puedes usar groupby!
```

```python
import pandas as pd
```

```python
f = open('Countries.csv','r')
countries = pd.read_csv(f)
countries
```

```python
# Encontrar el punto medio más alto de los países en la  no UE  maximo
countries.groupby('EU').max()
```

```python
# Encontrar el punto medio más alto de los países en la  no UE  promedio
countries.groupby('EU').mean()
```

```python
# Encontrar el punto medio más alto de los países en la  no UE 
countries[(countries.highest==max(countries[countries.EU == 'no'].highest))]
```

```python
# Encontrar el punto medio más alto de los países en la UE
countries[(countries.highest==max(countries[countries.EU == 'yes'].highest))]
```

```python
# Encontrar el punto medio más alto de los países  sin y con línea costera promedio
countries.groupby('coastline').mean()
```

```python
# Encontrar el punto medio más alto de los países  sin y con línea costera maxima
countries.groupby('coastline').max()
```

```python
# Encontrar el punto medio más alto de los países con línea costera
countries[(countries.highest==max(countries[countries.coastline == 'yes'].highest))]
```

```python
# Encontrar el punto medio más alto de los países  sin línea costera
countries[(countries.highest==max(countries[countries.coastline == 'no'].highest))]
```
