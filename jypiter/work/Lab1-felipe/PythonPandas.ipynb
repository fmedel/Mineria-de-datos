### Lectura desde archivos CSV en dataframe

```python
f = open('Cities.csv','rU')
cities = pd.read_csv(f)
cities
```

```python
# Número de Filas
cities.describe()
```

```python
# Ver las primeras filas
cities.head()
```

```python
# Las últimas 20 filas
cities.tail(20)
```

### Clasificación, selección de filas y columnas

```python
# Clasificación por país de forma ascendente, luego por temperatura descendente
cities.sort_values(by=['country','temperature'],ascending=[True,True])
```

```python
# Selección de una sola columna: devuelve una "serie"
cities.city
```

```python
# Selecciona múltiples columnas - retorna un dataframe
cities[['city','temperature']]
```

```python
# Selecciones filas basado en una condición
# Nota: no hay necesidad de hacer conversión de tipos - pandas infiere el tipo de la columna
cities[cities.longitude < 0]
```

```python
# Seleccione filas por rango
cities[200:]
# Ver cities[:8] y cities[200:]
```

```python
# Agrupar: seleccione filas y columnas, clasificandolos:
# Ciudades y longitudes de todas las ciudades con latitud > 50 y 
# temperatura > 9, ordenar por longitud
cities[(cities.latitude > 50) & (cities.temperature > 9)][['city','latitude','temperature','longitude']].sort_values(by='longitude')
```
