```python
import pandas as pd
```

```python
# 1.a) Leer los equipos en dataframes
csv_team = open('Teams.csv','r')
equipos = pd.read_csv(csv_team)
equipos
```

```python
# 1.b)Leer los jugadores en dataframes
csv_player = open('Players.csv','r')
jugadores = pd.read_csv(csv_player)
jugadores
```

<hr>
<h2>
 2) Qué jugador en un equipo con "ia" en el nombre, jugó menos de
200 minutos e hizo más de 100 pases? Imprima el apellido del jugador.
</h2>

```python

jugadores[(jugadores.team.str.contains('ia')) & (jugadores.minutes<200) & (jugadores.passes>100)].surname
```

<hr>
<h2>3.a) ¿Cuál es el promedio de pases de los defensas?</h2>

```python
jugadores[jugadores.position=='defender'].passes.mean()
```

<h2> 3.b) ¿Cuál es el promedio de pases de los delanteros?</h2>

```python
jugadores[jugadores.position=='forward'].passes.mean()
```

<hr><h2>4)¿Qué equipo tiene el mayor ratio de 'goalsFor' y 'goalsAgainst'? Imprimir sólo el nombre del equipo.</h2>

```python
equipos[(equipos.goalsFor==max(equipos.goalsFor)) & (equipos.goalsAgainst==max(equipos.goalsAgainst))].team
```

<h3><font color="red">Extra</font> ¿Qué equipo tiene el mayor ratio de 'goalsFor' o 'goalsAgainst'? Imprimir sólo el nombre del equipo.</h3>

```python

equipos[(equipos.goalsFor==max(equipos.goalsFor)) | (equipos.goalsAgainst==max(equipos.goalsAgainst))].team
```

<hr><h2>¿Cuántos jugadores que juegan en un equipo con ranking <10 jugaron
más de 350 minutos?</h2>

```python

jugadores_equipo = equipos.merge(jugadores)
jugadores_equipo[(jugadores_equipo.ranking<10)& (jugadores_equipo.minutes>350)].surname.count()
```

<h3><font color="red">Extra</font> ¿Cuántos jugadores que juegan en un equipo con ranking <10 jugaron más de 350 minutos? usando len() </h3>

```python
jugadores_equipo = equipos.merge(jugadores)
len(jugadores_equipo[(jugadores_equipo.ranking<10)& (jugadores_equipo.minutes>350)].surname)
```

```python
# Escribir un bucle que interactivamente le pide al usuario que introduzca un nombre de equipo.

# Si el equipo existe, liste todos los jugadores de ese equipo
# (con toda su información), ordenados por minutos descendentes.

# Si el equipo no existe, imprima "Este equipo no clasifico al mundial".
# Si se ingresa 'salir', termine el loop.  -> ok

# Recordatorio: Para leer una cadena del usuario en lugar de un número, utilice 
# raw_input () en lugar de input () ->ok

# Nota: Para probar si un valor v es (no) en una columna c de un dataframe D,
# use "v (not) in D ['c'].values"
```

```python
def pasar(datos_ingresado):
    dato_ingresado_final=dato_ingresado.lower()
    return dato_ingresado_final.capitalize()
def pasar_minus(datos_ingresado):
    return dato_ingresado.lower()
```

```python
dato_ingresado=""
while dato_ingresado != 'salir':
    dato_ingresado = input("Introduzca el nombre del equipo:") #python 3
    #dato_ingresado = raw_input("Introduzca el nombre del equipo:") #python 2
    if pasar_minus(dato_ingresado) == 'salir':
        print( "Saliendo....")#python3
        #print "Saliendo...." #python2
        break
    if equipos[equipos.team.str.contains(pasar(dato_ingresado))].team.count() == 0:
        print( "Este equipo no clasifico al mundial")#python3
        #print "Este equipo no clasifico al mundial" #python2
    else:
        print(jugadores[(jugadores.team == (pasar(dato_ingresado)))].sort_values(by='minutes',ascending=False))   #python3
        #print jugadores[(jugadores.team == (pasar(dato_ingresado)))].sort_values(by='minutes',ascending=False)   #python2 
        
```
