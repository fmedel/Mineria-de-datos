```python
#rango(de,hasta) 
for numero  in range(51472,52700):
    url = 'https://api.jikan.me/manga/'+str(numero)
    response = requests.get(url)
    if response.status_code == 200:
        results = response.json()
        archivo = open('datos.txt','a')
        for result in results:
            if result == "mal_id":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "title":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "type":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "volumes":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "chapters":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "published":
                try:
                    archivo.write(str(results[result]["from"])+"|")
                except:
                    archivo.write("none"+"|")
                try:
                    archivo.write(str(results[result]["to"])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "score": 
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "scored_by":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "rank":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "popularity":
                try:
                    archivo.write(str(results[result])+"|")
                except:
                    archivo.write("none"+"|")
            if result == "favorites":
                try:
                    archivo.write(str(results[result])+"|") 
                except:
                    archivo.write("none"+"|")
            if result == "status":
                try:
                    archivo.write(str(results[result])+"|") 
                except:
                    archivo.write("none"+"|")
            if result == "related":
                if results[result]:
                    archivo.write("si"+"|") 
                    try:
                        if results[result]["Adaptation"][0]:#revisar aqui
                             archivo.write("si"+"|") 
                        else:
                            archivo.write("no"+"|")
                    except:
                        archivo.write("no"+"|")
                else:
                    archivo.write("no"+"|") 
                    archivo.write("no"+"|")     
            if result == "author":
                try:
                    archivo.write(results[result][0]["name"]+"|")
                except:
                    archivo.write("none"+"|")
            if result == "serialization":
                try:
                    if results[result]:
                         archivo.write(str(results[result][0]["name"])+"\n") 
                    else:
                        archivo.write("none"+"\n")
                except:
                    archivo.write("none"+"\n")
            if result == "genre":
                generos=""
                for genero in results[result]:
                    try:
                        if(generos==""):
                            generos=genero["name"]
                        else:
                            generos=generos+","+genero["name"]
                    except:
                        archivo.write("none"+"|")
                if generos=="":
                    archivo.write("none|") 
                else:
                    archivo.write(generos+"|")
                
        archivo.close()
    else:
        print("Error code %s" % response.status_code)

print("termino")
```
